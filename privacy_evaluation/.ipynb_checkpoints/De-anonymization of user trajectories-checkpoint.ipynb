```python
import pandas as pd
import numpy as np
from collections import Counter
import math
from grid import location_to_cell_id
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_queries = pd.read_csv("queries.csv", sep = " ")
df_pois = pd.read_csv("pois.csv", sep = " ")
```

```python
# 200 users over 20 days
# monday: between 0 and 24 %7 OR 
# tuesday: between 24 and 48 % 7
# ...
# here we consider the first Monday, Tuesday, ... for simplicity
def this_day(start, end, row):
    time = row['timestamp']
    if (start <= time and time < end):
        return True
    else:
        return False
    
def this_time(start, end, row):
    time = row['timestamp']
    if (start <= time%24 and time%24 <= end):
        return True
    else:
        return False
    
def is_monday(row):
    return this_day(0,24,row)
def is_tuesday(row):
    return this_day(24,48,row)
def is_wednesday(row):
    return this_day(48,72,row)
def is_thursday(row):
    return this_day(72,96,row)
def is_friday(row):
    return this_day(96,120,row)
def is_saterday(row):
    return this_day(120,148,row)
def is_sunday(row):
    return this_day(148,172,row)

def is_weekend(row):
    return is_saterday(row) or is_sunday(row)
def is_weekday(row):
    return not is_weekend(row)

def is_work_time(row):
    return this_time(8,11,row) or this_time(14,17,row)
def is_lunch_time(row):
    return this_time(11,14,row)
def is_night(row):
    return this_time(0,6,row)
```

#### people they might have eaten with 

```python
user_lunch = df_user[df_user.apply(is_lunch_time, axis=1)]
user_lunch["gid"] = user_lunch.apply(lambda r: location_to_cell_id(r['lat'], r['lon']), axis=1)
user_gids_time = user_lunch[["gid","timestamp"]]
df_queries_gid = df_queries
df_queries_gid["gid"] = df_queries.apply(lambda r: location_to_cell_id(r['lat'], r['lon']), axis=1)
df_queries_gid = df_queries_gid[["ip_address","timestamp","gid"]]
threshold = 0.05 # less than 3 minutes at same grid location

sets = []

for idx,row in user_gids_time.iterrows():
    gid = row["gid"]
    timestamp = row["timestamp"]
    user_colleagues = df_queries_gid[df_queries_gid.apply(lambda q: q["gid"] == gid and abs(q["timestamp"]-timestamp)<threshold, axis=1)]
    user_colleagues = user_colleagues[["ip_address"]].drop_duplicates()
    set_colleagues = set(user_colleagues["ip_address"].unique())
    print(set_colleagues)
    sets.append(set_colleagues)
```

#### people that where around the same location during the night: potentially the partner or member of family or close person.

```python
def euclid_dist(lat_1,lat_2,lon_1,lon_2):
    return math.sqrt((lat_1-lat_2)**2 + (lon_1-lon_2)**2)
```

```python
user_night = df_user[df_user.apply(is_night, axis=1)][["lat","lon"]]
dist_threshold = 0.00001

sets = []

for idx,row in user_night.iterrows():
    lat = row["lat"]
    lon = row["lon"]
    night_network = df_queries[df_queries.apply(lambda q: euclid_dist(q["lat"],lat,q["lon"],lon)<dist_threshold, axis=1)]
    night_network = night_network[["ip_address"]].drop_duplicates()
    set_night_nw = set(night_network["ip_address"].unique())
    print(set_night_nw)
    sets.append(set_night_nw)
```

## Defense

 - add dummy queries with same IP same timestamp (should be subscribed)
 
 
 - (not considered in this part) unlink IP and queries: send random and different IPs at each query from 1 same user (user needs to prove credentials for each query) --> LSP cannot link queries from a same user together and cannot link neither of the queries to the corresponding user (may still be possible if user is the only one in the grid cell and LSP somehow knows who is in this specific cell).
     - LSP still has access to users subscriptions and thus to their interests: cannot associate queries to a single user but to a set (potentially very small) of users with same interests. E.g. if k users have the same subscribtion set for all possible sets, then k anonymity is provided.
     - does not hurt utility

### Solution implementation

```python
df_user
```

```python
bob_subscriptions = set(df_user["poi_type_query"].unique())
bob_subscriptions
```

```python
# for all rows, add 4 other rows corresponding to other POI types subscriptions
dummies = pd.DataFrame({'ip_address': [], 'lat': [], 'lon': [], 'timestamp': [], 'poi_type_query': []})
for idx,row in df_user.iterrows():
    poi = row["poi_type_query"]
    lat = row["lat"]
    lon = row["lon"]
    for p in bob_subscriptions.difference({poi}):
        dummy = pd.DataFrame({'ip_address': [row['ip_address']], 'lat': [row['lat']], 'lon': [row['lon']], 'timestamp': [row['timestamp']], 'poi_type_query': [p]})
        dummies = dummies.append(dummy, ignore_index=True)
```

```python
df_user_w_dummies = df_user.append(dummies, ignore_index=True)
df_user_w_dummies
```
